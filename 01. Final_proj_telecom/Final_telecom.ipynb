### Проверка и приведение данных к нужному типу.

#### datetime

Значения даты содержатся в столбцах 'BeginDate' и 'EndDate' (Заключение и Окончание договора)  
В столбце 'EndDate' есть знаачения 'no' - это значит что на момент выгрузки данных (01.02.2020) договор еще действителен.  
Чтобы рассчитать срок действия договора заменим значение 'no' на дату выгрузки данных

```python
# Дата выгрузки данных - EndDate для действующих договоров

df.loc[df['EndDate']=='no', 'EndDate'] = '2020-02-01'
```

```python
# Установим формат datettime для прирзнаков с датой

for col in ['BeginDate', 'EndDate']:
    df[col] = pd.to_datetime(df[col], format='%Y-%m-%d')
```

### Статистический тест

**Нулевая гипотеза:**  
Средние ежемесячные платежи у оставшихся и ушедших клиентов равны  

**Альтернативная гипотеза:**  
Средние ежемесячные платежи у оставшихся и ушедших клиентов не равны

```python
# Проведем статистический тест на равенство средних ежемесяных платежей

alpha = .01 # критический уровень статистичесой значимости
results = st.ttest_ind(remain_clients['MonthlyCharges'], leave_clients['MonthlyCharges']) 
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Статистический тест позволяет утверждать, что с большой долей вероятности Средние ежемесячные платежи у оставшихся и ушедших клиентов не равны

**Нулевая гипотеза:**  
Средние общие расходы у оставшихся и ушедших клиентов равны  

**Альтернативная гипотеза:**  
Средние общие расходы у оставшихся и ушедших клиентов не равны

```python
# Проведем статистический тест на равенство средних общих расходов

alpha = .01 # критический уровень статистичесой значимости
results = st.ttest_ind(remain_clients['TotalCharges'], leave_clients['TotalCharges']) 
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Статистический тест позволяет утверждать, что с большой долей вероятности Средние общие расходы у оставшихся и ушедших клиентов не равны

## Обучение моделей и Кроссвалидация (GridSearchCV)

<font size="3"> 
    
- Проверим на адекватность с помощью **DummyClassifier**

- Добавим визуализацию (Матрица ошибок и ROC_curve)

### Обучение выбранных моделей

```python
# Добавим модели и лучшие гиперпараметры в списки

model_list = [
    model_logistic,
    model_tree,
    model_forest
]
best_params_list = [
    logistic_best_params,
    tree_best_params,
    forest_best_params]

# Список с названиями моделей
name_list = ['LogisticRegression', 'DecisionTree', 'RandomForest']
```

```python
# Выберем лучшие модели с помощью CV (Быстрые модели)

grid_best_score_list = []
grid_model_list = []
predictions = []
timing = []
timing_pred = []
f1_list = []
accuracy_list = []


for n, model in enumerate(model_list):
    
    grid_model = GridSearchCV(model, best_params_list[n], scoring = 'roc_auc')
    start = time.time() # Засекаем время обучения
    grid_model.fit(features_train, target_train)
    end = time.time() # Стоп
    timing.append(end-start)
    
    # Время предсказания
    start_pred = time.time()
    grid_model_predictions = cross_val_predict(model, features_train, target_train, cv=5)
    end_pred = time.time()
    predictions.append(grid_model_predictions)
    timing_pred.append(end_pred - start_pred)
    print()
     # Вероятности
    probabilities = cross_val_predict(model, features_train, target_train, cv=5, method='predict_proba')
    probabilities_ones = probabilities[:,1]
    
    # Рассчет метрик 
    auc_roc_train = grid_model.best_score_
    fpr, tpr, thresholds = roc_curve(target_train, probabilities_ones)
    
    f1 = f1_score(target_train, grid_model_predictions, zero_division=0)
    f1_list.append(f1)
    accuracy = accuracy_score(target_train, grid_model_predictions)
    accuracy_list.append(accuracy)
    
    print(name_list[n])
    print(f'Время обучения: {end - start}')
    print(f'Время предсказания: {end_pred - start_pred}')
    print(f'AUC_ROC: {grid_model.best_score_}')
    print(f'f1-мера: {f1}')
    print(f'Accuracy: {accuracy}')
    
    # Матрица Ошибок
    fig, ax = plt.subplots(figsize = (4,3))
    cm = confusion_matrix(target_train, predictions[n], labels = (0,1))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=(0,1))
    disp.plot(cmap=plt.cm.Blues, ax = ax)

    plt.xlabel('Предсказания')
    plt.ylabel('Ответы')
    plt.show()
    
    # Графики ROC-curve
    fig,ax = plt.subplots(figsize = (4,3))
    # ROC-curve Случайной модели
    ax.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1),label = 'Случайная модель',ls = '--')
    
    # ROC-curve лучших моделей
    ax.plot(fpr,tpr,
            label = f'{name_list[n]})')
    ax.legend()
    ax.set_title(f'ROC-curve:, {name_list[n]}')
    ax.set_xlabel('FPR')
    ax.set_ylabel('TPR')
    ax.set_xlim((0,1))
    ax.set_ylim((0,1))
    plt.show()



    grid_best_score_list.append(grid_model.best_score_)
    grid_model_list.append(grid_model.best_estimator_)
    
```

```python
# Добавим модели и лучшие гиперпараметры в списки

boosting_list = [model_lgbm, cat_models]
boosting_best_params = [
    lgbm_best_params,
    cat_best_params
]

# Список с названиями моделей
name_boosting = ['LGBM', 'CatBoost']
```

```python
# Выберем лучшие модели с помощью CV (boosting)

grid_best_score_boost = []
grid_model_boost = []


for n, model in enumerate(boosting_list):
    
    grid_model = GridSearchCV(model, boosting_best_params[n], scoring = 'roc_auc')
    start = time.time() # Засекаем время обучения
    grid_model.fit(features_train, target_train)
    end = time.time() # Стоп
    timing.append(end-start)
    
    # Время предсказания
    start_pred = time.time()
    grid_model_predictions = cross_val_predict(model, features_train, target_train, cv=5)
    predictions.append(grid_model_predictions)
    end_pred = time.time()
    timing_pred.append(end_pred - start_pred)
    print()
    
    # Вероятности
    probabilities = cross_val_predict(model, features_train, target_train, cv=5, method='predict_proba')
    probabilities_ones = probabilities[:,1]
    
    # Рассчет метрик 
    auc_roc_train = grid_model.best_score_
    fpr, tpr, thresholds = roc_curve(target_train, probabilities_ones)
    
    f1 = f1_score(target_train, grid_model_predictions, zero_division=0)
    f1_list.append(f1)
    accuracy = accuracy_score(target_train, grid_model_predictions)
    accuracy_list.append(accuracy)
    
    print(name_boosting[n])
    print(f'Время обучения: {end - start}')
    print(f'Время предсказания: {end_pred - start_pred}')
    print(f'AUC_ROC: {grid_model.best_score_}')
    print(f'f1-мера: {f1}')
    print(f'Accuracy: {accuracy}')
    
    # График Матрицы ошибок
    fig, ax = plt.subplots(figsize = (4,3))
    cm = confusion_matrix(target_train, predictions[n], labels = (0,1))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=(0,1))
    disp.plot(cmap=plt.cm.Blues, ax = ax)

    plt.xlabel('Предсказания')
    plt.ylabel('Ответы')
    plt.show()
 
    # Графики ROC-curve
    fig,ax = plt.subplots(figsize = (4,3))
    # ROC-curve Случайной модели
    ax.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1),label = 'Случайная модель',ls = '--')
    
    # ROC-curve лучших моделей
    ax.plot(fpr,tpr,
            label = f'{name_boosting[n]})')
    ax.legend()
    ax.set_title(f'ROC-curve:, {name_boosting[n]}')
    ax.set_xlabel('FPR')
    ax.set_ylabel('TPR')
    ax.set_xlim((0,1))
    ax.set_ylim((0,1))
    plt.show()

    grid_best_score_boost.append(grid_model.best_score_)
    grid_model_boost.append(grid_model.best_estimator_)
```

```python
# Добавим значения в общий список

grid_best_score_list.extend(grid_best_score_boost)
grid_model_list.extend(grid_model_boost)
name_list.extend(name_boosting)

time_models = np.array(timing)
time_pred = np.array(timing_pred)
```

```python
for n, value in enumerate(grid_best_score_list):
    print()
    print(f'ROC_AUC модели: {name_list[n]}: {value:.03f}')
    if value > 0.85:    
        print(f'Модель {name_list[n]}, подходит под условие проекта')
    elif value <= 0.85:
        print(f'Модель {name_list[n]} не подходит')
        
    print(f'Время обучения: {time_models[n]:.03f} sec.')
    print(f'Время Предсказания: {time_pred[n]:.03f} sec.')
```

```python
# Для Наглядности соберем показатели в табличку

report = pd.DataFrame(columns = name_list, index = ['ROC_AUC', 'F1-мера', 'Accuracy', 'Скорость Обучения (sec)', 'Скорость Прогноза (sec)'])
report.loc['ROC_AUC',:] = grid_best_score_list
report.loc['F1-мера',:] = f1_list
report.loc['Accuracy',:] = accuracy_list
report.loc['Скорость Обучения (sec)',:] = time_models
report.loc['Скорость Прогноза (sec)',:] = time_pred
report
```

1. В этом проекте выполняли задачу бинарной классификации для Телеком-компании, чтобы прогнозировать отток клиентов.  
Если выяснится, что пользователь планирует уйти, компания предложит такому клиенту промокоды и специальные условия.
  
    
Выполнили задачи проекта:    
- Провести исследовательский анализ данных. Выяснить какие факторы наболее всего влияют на принятие клиентом решения прекратить пользоваться услугами телеком-компании.  
- Обучить разные модели ML для прогноза оттока клиентов компании. Выбрать модель лучшего качества

BeginDate, duration, TotalCharges, Partner, MonthlyCharges те признаки, которые наиболее взаимосвязаны с признаком leave (Ушел клиент или нет)

Рост количества подписанных договоров наблюдается в 2014 году  также В конце 2019 к началу 2020 гг.

Статистический тест позволяет утверждать, что с большой долей вероятности Средние ежемесячные платежи и Средний общий чек у оставшихся и ушедших клиентов разные
